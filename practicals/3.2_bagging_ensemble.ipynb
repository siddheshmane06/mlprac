```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import warnings
warnings.filterwarnings('ignore')
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

df.sample(4)
```

```python
df.shape
```

```python
sample_1 = df.sample(200, replace = True)
sample_2 = df.sample(200, replace = True)
sample_3 = df.sample(200, replace = True)
```

```python
sample_1.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
base_model_1 = DecisionTreeClassifier(max_depth=2)
base_model_2 = DecisionTreeClassifier(max_depth=2)
base_model_3 = DecisionTreeClassifier(max_depth=2)
```

```python
base_model_1
```

```python
sample_1.iloc[:,:30]
```

```python
base_model_1.fit(sample_1.iloc[:,:30],sample_1.iloc[:,-1])
base_model_2.fit(sample_2.iloc[:,:30],sample_2.iloc[:,-1])
base_model_3.fit(sample_3.iloc[:,:30],sample_3.iloc[:,-1])
```

```python
for i in df.sample():
    print(i)
```

```python
df.sample().values
```

```python
df.drop(df['target'].values)
```

```python
sample_ = df.iloc[:,:30].sample(1).values
sample_
```

```python
#base_model_1.predict(df.iloc[:,:30].sample(1).values)
```

```python
model1 = base_model_1.predict(sample_)
model1
```

```python
model2 = base_model_2.predict(sample_)
model2
```

```python
model3 = base_model_3.predict(sample_)
model3
```

```python
predicted_op = np.concatenate((model1, model2, model3))
predicted_op
```

```python
from statistics import mode, mean
output_mode = mode(predicted_op)
output_mean = mean(predicted_op)
```
